## Подготовка

Подгтовим данные так же как и лдя размера 1000-1100, чтобы были строки с результатами в разных файлах для разных типов.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

simple_results = []
with open('simple_results_1000_2000.txt') as fin:
    for line in fin:
        simple_results.append([float(n) for n in line.split()])
        
block_results = []
with open('block_results_1000_2000.txt') as fin:
    for line in fin:
        block_results.append([float(n) for n in line.split()])
        
swapped_results = []
with open('swapped_results_1000_2000.txt') as fin:
    for line in fin:
        swapped_results.append([float(n) for n in line.split()])

plt.rcParams["figure.figsize"] = (15, 5)
```

# Простое умножение

```python
for results in simple_results:
    plt.plot(xrange(1000, 2000, 10), results)

plt.xticks(np.arange(1000, 2000, 50))
plt.show()
```

# Блочное множение

```python
for results in block_results:
    plt.plot(xrange(1000, 2000, 10), results)

plt.xticks(np.arange(1000, 2000, 50))
plt.show()
```

Можно заметить, что график начинает "трясти" при росте размеров матриц. 

# Свапнутое умножение

```python
for results in swapped_results:
    plt.plot(xrange(1000, 2000, 10), results)

plt.xticks(np.arange(1000, 2000, 50))
plt.show()
```

Тут все стабильно, как на меньших размерах. 

# Сравнение трех алгоритмов вместе

```python
import matplotlib.pyplot as plt

average_simple = np.sum(simple_results, axis=0) / len(simple_results)
average_block = np.sum(block_results, axis=0) / len(block_results)
average_swapped = np.sum(swapped_results, axis=0) / len(swapped_results)

plt.plot(xrange(1000, 2000, 10), average_simple)
plt.plot(xrange(1000, 2000, 10), average_block)
plt.plot(xrange(1000, 2000, 10), average_swapped)

plt.xticks(np.arange(1000, 2000, 50))
plt.show()
```
